```python
import spacy
from docling.document_converter import DocumentConverter
from google.colab import files
import os

# ✅ Step 1: Upload ANY document (PDF, DOCX, TXT)
uploaded = files.upload()
file_path = list(uploaded.keys())[0]  # Get uploaded filename

# ✅ Step 2: Extract Text
converter = DocumentConverter()

file_ext = os.path.splitext(file_path)[1].lower()

if file_ext == ".txt":
    print("📌 TXT detected → Reading text directly")
    with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
        extracted_text = f.read()
else:
    print("📌 Converting using Docling...")
    result = converter.convert(file_path)
    extracted_text = result.document.export_to_text()

print("\n📄 Extracted Text Preview:\n")
print(extracted_text[:500])  # print first 500 characters

# ✅ Step 3: spaCy NER
nlp = spacy.load("en_core_web_sm")
doc = nlp(extracted_text)

# ✅ Step 4: Display extracted named entities
print("\n✅ Named Entities Detected:\n")
if len(doc.ents) == 0:
    print("⚠️ No entities detected!")
else:
    for ent in doc.ents:
        print(f"{ent.text} → {ent.label_}")


```

```python
from spacy import displacy

html_output = displacy.render(doc, style="ent", page=True, jupyter=False)

output_file = "entity_visualization.html"
with open(output_file, "w", encoding="utf-8") as f:
    f.write(html_output)

print("✅ Entity visualization saved as:", output_file)
```
