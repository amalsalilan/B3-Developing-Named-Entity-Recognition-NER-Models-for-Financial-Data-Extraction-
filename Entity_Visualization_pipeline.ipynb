```python
from google.colab import files
from docling.document_converter import DocumentConverter
import spacy
from spacy import displacy
from IPython.display import HTML, display

# ---- File Upload ----
uploaded = files.upload()
file_path = list(uploaded.keys())[0]
file_path = f"/content/{file_path}"

# ---- 1. Document â†’ Plain Text ----
def extract_text(file_path):
    converter = DocumentConverter()
    result = converter.convert(file_path)
    text = result.document.export_to_text()  # âœ… Correct method
    return text

# ---- 2. NER Model ----
def perform_ner(text):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)
    return doc

# ---- 3. Visualization ----
def visualize_entities(doc):
    html = displacy.render(doc, style="ent", page=True)
    display(HTML(html))  # Displays in Colab cell

# ---- 4. Main Pipeline ----
def main(file_path):
    print("ğŸ”¹ Extracting text...")
    text = extract_text(file_path)
    print("âœ… Text extracted successfully!\n")

    print("ğŸ”¹ Performing NER...")
    doc = perform_ner(text)
    print("âœ… NER complete! Visualizing entities...\n")

    visualize_entities(doc)

# ---- Run the pipeline ----
main(file_path)
```
