# Task
Analyze the sentiment of a financial PDF document using FinBERT, displaying color-coded results and a summary of the sentiment distribution.

## Upload PDF File

### Subtask:
Provide a mechanism for the user to upload the financial PDF document. The agent will prompt for the file path or a file upload widget.


**Reasoning**:
The user needs to upload a PDF file using a file upload widget provided by Google Colab. This code block imports the necessary module, initiates the upload widget, and then extracts the filename of the uploaded document.



```python
from google.colab import files

uploaded = files.upload()

# Extract the filename
for filename in uploaded.keys():
  pdf_file_path = filename
  print(f"Uploaded file: {pdf_file_path}")
```

```python
import nltk
# Download both punkt and punkt_tab for full sentence tokenization support
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# ============================================
# STEP 1: Extract Text from PDF
# ============================================
from docling.document_converter import DocumentConverter

converter = DocumentConverter()
result = converter.convert(pdf_file_path)
text_content = result.document.export_to_markdown()

# ============================================
# STEP 2: Clean & Prepare Text
# ============================================
import re
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

def clean_text(text):
    # Remove markdown artifacts and non-printable characters
    text = re.sub(r'\*\*', '', text)
    text = re.sub(r'[_`#>-]', '', text)
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    return text.strip()

cleaned_text = clean_text(text_content)

# Sentence tokenization
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(cleaned_text)
print(f"📜 Extracted {len(sentences)} sentences for sentiment analysis.")

# ============================================
# STEP 3: Perform Sentence-Level Sentiment Analysis
# ============================================
from transformers import pipeline

sentiment_pipeline = pipeline("text-classification", model="ProsusAI/finbert")

results = []
for i, sentence in enumerate(sentences):
    sentiment = sentiment_pipeline(sentence[:512])[0]
    results.append({
        "sentence_index": i + 1,
        "text": sentence,
        "label": sentiment['label'],
        "score": sentiment['score']
    })

print("✅ Sentence-level sentiment analysis completed!")

# ============================================
# STEP 4: Define Enhanced Color Scheme
# ============================================
def sentiment_color(label, score):
    label = label.lower()
    if label == "positive":
        if score >= 0.85:
            return "#00B050"  # very positive (deep green)
        elif score >= 0.65:
            return "#92D050"  # positive (light green)
    elif label == "negative":
        if score < 0.25:
            return "#C00000"  # very negative (dark red)
        elif score < 0.45:
            return "#F8696B"  # negative (coral)
    return "#FFD966"  # neutral (warm yellow)

# ============================================
# STEP 5: Generate Color-Coded Output
# ============================================
from IPython.display import display, HTML

html_output = """
<h3 style='font-family:Arial; color:#333;'>📊 Financial Sentiment Analysis Report</h3>
<div style='font-family:Arial;'>
"""

pos, neu, neg = 0, 0, 0

for r in results:
    color = sentiment_color(r['label'], r['score'])
    html_output += f"""
    <div style='background-color:{color}; padding:12px; border-radius:10px; margin:8px 0; box-shadow:0 2px 5px rgba(0,0,0,0.1);'>
        <b>Sentence {r['sentence_index']}</b> —
        <b style='color:#333'>{r['label'].capitalize()}</b>
        (Score: {r['score']:.3f})<br>
        <span style='font-size:15px; color:#222;'>{r['text']}</span>
    </div>
    """
    if "positive" in r['label'].lower():
        pos += 1
    elif "negative" in r['label'].lower():
        neg += 1
    else:
        neu += 1

# ============================================
# STEP 6: Add Summary Section
# ============================================
total = pos + neu + neg
pos_pct = (pos/total)*100 if total > 0 else 0
neu_pct = (neu/total)*100 if total > 0 else 0
neg_pct = (neg/total)*100 if total > 0 else 0

html_output += f"""
<hr>
<h4 style='color:#333;'>📈 Sentiment Summary</h4>
<div style='display:flex; gap:10px; font-size:15px;'>
    <div style='background:#00B050; color:white; padding:6px 10px; border-radius:6px;'>Positive: {pos} ({pos_pct:.1f}%)</div>
    <div style='background:#FFD966; color:black; padding:6px 10px; border-radius:6px;'>Neutral: {neu} ({neu_pct:.1f}%)</div>
    <div style='background:#C00000; color:white; padding:6px 10px; border-radius:6px;'>Negative: {neg} ({neg_pct:.1f}%)</div>
</div>
</div>
"""

display(HTML(html_output))
print("✅ Displayed enhanced color-coded sentence-level sentiment analysis (with improved colors)!")
```

```python
# ============================================
# FINBERT SENTIMENT ANALYSIS (Ordered Output with Visible Colors)
# ============================================

import re
import nltk
import torch
import numpy as np
from nltk.tokenize import sent_tokenize
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from IPython.display import display, HTML

nltk.download('punkt')
nltk.download('punkt_tab')

# ---------- STEP 1: Clean Text ----------
def clean_text(text: str) -> str:
    text = re.sub(r'\*\*', '', text)
    text = re.sub(r'\*', '', text)
    text = re.sub(r'[_`#>-]', '', text)
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    return text.strip()

# cleaned_text = clean_text(text_content)
sentences = sent_tokenize(cleaned_text)
print(f"📜 Extracted {len(sentences)} sentences for sentiment analysis.")

# ---------- STEP 2: Run FinBERT ----------
tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
model.eval()
id2label = model.config.id2label

results = []
for i, sentence in enumerate(sentences):
    if not sentence.strip():
        continue
    inputs = tokenizer(sentence, return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        logits = model(**inputs).logits[0].detach().numpy()
        probs = np.exp(logits) / np.sum(np.exp(logits))

    prob_dict = {id2label[j].lower(): probs[j] for j in range(len(probs))}
    label = max(prob_dict, key=prob_dict.get)
    results.append({
        "idx": i + 1,
        "text": sentence.strip(),
        "probs": prob_dict,
        "label": label,
        "score": prob_dict[label]
    })

print("✅ FinBERT sentiment probabilities calculated for all sentences!")

# ---------- STEP 3: Map Severity ----------
def sentiment_category(probs):
    pos, neu, neg = probs.get("positive", 0), probs.get("neutral", 0), probs.get("negative", 0)
    if pos > 0.45 and pos > neg:
        return "very_positive" if pos >= 0.75 else "positive"
    elif neg > 0.40 and neg > pos:
        return "very_negative" if neg >= 0.70 else "negative"
    return "neutral"

for r in results:
    r["category"] = sentiment_category(r["probs"])

# ---------- STEP 4: Define Colors ----------
CATEGORY_META = {
    "very_positive": {"label": "Very Positive", "bg": "#00B050", "fg": "#FFFFFF"},
    "positive": {"label": "Positive", "bg": "#90EE90", "fg": "#000000"},
    "neutral": {"label": "Neutral", "bg": "#FFF59D", "fg": "#000000"},
    "negative": {"label": "Negative", "bg": "#FF6F61", "fg": "#FFFFFF"},
    "very_negative": {"label": "Very Negative", "bg": "#C00000", "fg": "#FFFFFF"},
}

# ---------- STEP 5: Generate Ordered Visualization ----------
html = """
<style>
body { font-family: 'Inter', Arial, sans-serif; color:#111; }
.analysis-container { margin-top:15px; }
.card {
  border-radius: 10px; margin:10px 0; padding:12px 14px;
  box-shadow: 0 1px 5px rgba(0,0,0,0.15);
}
.badge {
  float:right; padding:3px 8px; border-radius:6px; font-weight:600;
  font-size:12px;
}
small { color:#333; font-size:13px; }
@media (prefers-color-scheme: dark) {
  body { background:#1e1e1e; color:#eee; }
  .card { background:#2a2a2a; box-shadow:0 1px 5px rgba(255,255,255,0.08); }
  small { color:#ccc; }
}
</style>
<div class='analysis-container'>
<h3>📊 Financial Sentiment Analysis — Ordered by Document Flow</h3>
"""

for r in results:
    cat = CATEGORY_META[r["category"]]
    bg, fg = cat["bg"], cat["fg"]
    html += f"""
    <div class='card' style='background:{bg}1A;'>  <!-- light transparent background -->
        <b>Sentence {r['idx']}</b>
        <span class='badge' style='background:{bg}; color:{fg};'>{cat['label']}</span><br>
        <div style='color:{fg if bg != "#FFF59D" else "#000"}; font-size:15px;'>{r['text']}</div>
        <div style='margin-top:5px;'>
            <small>P(Positive): {r['probs'].get('positive',0):.2f} |
            P(Neutral): {r['probs'].get('neutral',0):.2f} |
            P(Negative): {r['probs'].get('negative',0):.2f}</small>
        </div>
    </div>
    """

html += "<div style='margin-top:15px; color:#666; font-size:13px;'>Generated by FinBERT — Sentence-level probabilities visualized inline.</div></div>"

display(HTML(html))
print("✅ Displayed ordered sentiment visualization with green-yellow-red theme!")
```
