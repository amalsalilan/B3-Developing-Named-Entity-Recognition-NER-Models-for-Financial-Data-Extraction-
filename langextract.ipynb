```python
import langextract as lx
import textwrap

# 1. Define the extraction prompt
prompt = textwrap.dedent("""\
    Extract the termination clause from the document using the exact text.
""")

# 2. Provide new finance-related examples
examples = [
    lx.data.ExampleData(
        text="Either party may terminate this Investment Management Agreement by giving ninety (90) days’ prior written notice to the other party.",
        extractions=[
            lx.data.Extraction(
                extraction_class="termination_clause",
                extraction_text="Either party may terminate this Investment Management Agreement by giving ninety (90) days’ prior written notice to the other party.",
                attributes={
                    "type": "Termination with Notice",
                    "triggering_party": "Either Party",
                    "notice_period": "90 days",
                    "condition": "none specified (general termination)",
                    "effect": "Agreement termination after notice period"
                }
            )
        ]
    ),
    lx.data.ExampleData(
        text="The Client may terminate this Agreement immediately if the Manager loses its investment license or becomes insolvent.",
        extractions=[
            lx.data.Extraction(
                extraction_class="termination_clause",
                extraction_text="The Client may terminate this Agreement immediately if the Manager loses its investment license or becomes insolvent.",
                attributes={
                    "type": "Termination for Cause",
                    "triggering_party": "Client",
                    "notice_period": "Immediate",
                    "condition": "loss of license or insolvency",
                    "effect": "Immediate termination"
                }
            )
        ]
    )
]
```

```python
input_text = """ INVESTMENT MANAGEMENT AGREEMENT

This Investment Management Agreement (“Agreement”) is entered into between AlphaWealth Advisors Pvt. Ltd. (“Manager”) and Sunrise Mutual Fund (“Client”) as of April 2, 2025.

1. Scope of Services
The Manager shall manage the Client’s investment portfolio in accordance with SEBI regulations and the investment policy guidelines provided by the Client.

2. Fees and Expenses
The Client agrees to pay a management fee of 1.25% per annum of assets under management, payable quarterly in arrears.

3. Confidentiality
Both parties agree to maintain strict confidentiality of investment data, reports, and strategies.

4. Termination
Either party may terminate this Agreement by giving ninety (90) days’ written notice to the other party.
The Client may terminate immediately if the Manager loses its investment license, violates SEBI regulations, or becomes insolvent.
Upon termination, the Manager shall transfer all assets, records, and reports to the Client within fifteen (15) business days and shall be paid fees accrued up to the termination date.

5. Governing Law
This Agreement shall be governed by and construed in accordance with the laws of India.

IN WITNESS WHEREOF, the parties have executed this Agreement as of the Effective Date.
"""
```

```python
# Run the extraction
result = lx.extract(
    text_or_documents=input_text,
    prompt_description=prompt,
    examples=examples,
    api_key="your_api_key",
    model_id="gemini-2.5-flash",
)
```

```python
# Save the results to a JSONL file
lx.io.save_annotated_documents([result], output_name="extraction_results.jsonl", output_dir=".")

# Generate the visualization from the file
html_content = lx.visualize("extraction_results.jsonl")
with open("visualization.html", "w") as f:
    if hasattr(html_content, 'data'):
        f.write(html_content.data)  # For Jupyter/Colab
    else:
        f.write(html_content)
```
